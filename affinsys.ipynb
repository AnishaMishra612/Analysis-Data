# Lets begin with Data Preparation and Cleaning:


```python
len(d1['InvoiceDate'].unique())
```

```python
# Converting the InvoiceDate into datetime object to analyse using dates
d1['InvoiceDate']=pd.to_datetime(d1['InvoiceDate'])
d1
```

### Adding separate column for Year and Month

```python
d1['year'] = d1['InvoiceDate'].dt.year
d1
```

```python
d1['month']=d1['InvoiceDate'].dt.month
d1
```

```python
d1['day']=d1['InvoiceDate'].dt.dayofweek
d1
```

```python
d1['Hour']=d1['InvoiceDate'].dt.hour
d1
```

```python
# To check rows will Null description
des_null = d1[d1['Description'].isna()]
des_null
```

```python
# To check rows with Null Customer ID
null_cust=d1[d1['CustomerID'].isna()]
null_cust
```

### FROM above two null column data, namely description and customerID, we see that by removing null customerID records, we can remove the null description records too.

## Hence dropping the null valued Rows frorm our table:


```python
d2 = d1.dropna()
```

```python
d2.isnull().sum()
# Hence null rows have been removed
```

```python
d2.info()
```

```python
d2.shape
```

### Check for duplicate rows and remove them:


```python
duplicate_rows = d2[d2.duplicated()]
print('number of duplicate rows: ', duplicate_rows.shape)
```

```python
d2=d2.drop_duplicates()
```

```python
d2
```

```python
duplicate_rows = d2[d2.duplicated()]
print('number of duplicate rows: ', duplicate_rows.shape)
```

## Removing negative quantity values for the new dataframe

```python
d2= d2[d2.Quantity > 0]
d2
```

# Exploratory data analysis

```python
##To find the date range of data
yearly=d2.groupby(['year','month']).sum()
yearly
```

#### Here we see that the Data is from 2010 December to 2011 December

```python
high_country=d2.groupby(['Country'])['InvoiceNo'].count().sort_values(ascending=False)[:10]
high_country
```

### Above plot: Top 10 countries having the maximum numbers of order placed



```python
# For each country showing the number of orders placed except UK
Country_wise=d2.groupby(['Country'])['InvoiceNo'].count().sort_values(ascending=False)
del Country_wise['United Kingdom']
Country_wise
```

#### Plot showing count of orders for each country except UK

## Top 10 countries producing Maximum Revenue 

```python
#Adding a column Total Amount:
d2['Total_amount'] = d2['Quantity'] * d2['UnitPrice']
d2
```

```python
high_amount=d2.groupby(['Country'])['Total_amount'].sum().sort_values(ascending=False)[:10]
high_amount
```

```python
high_amount.plot(kind="bar", figsize=(10, 10)).set_title("Top 10 countries with highest amount")
```

## Top 10 items producing Max amount

```python
high_Item_rev=d2.groupby(['Description'])['Total_amount'].sum().sort_values(ascending=False)[:10]
high_Item_rev
```

```python
high_Item_rev.plot(kind="bar", figsize=(10, 10)).set_title("Top 10 items producing highest amount")
```

```python
monthly=d2.groupby(['year','month']).agg({'Total_amount':'sum','InvoiceNo':'count'}).reset_index()
monthly
```

```python
monthly_2011=monthly[(monthly['year']==2011)]
monthly_2011
```

```python
monthly_2011.plot(x='month',y='Total_amount',kind="line", figsize=(10, 10)).set_title("Monthly amount for year 2011")
```

#### Above plot: Monthly revenue for year 2011
